```
#importing pandas 
import pandas as pd

#importing matplotlib for plotting later.
import matplotlib.pyplot as plt

# Reading in current data from coinmarketcap API
crypto = pd.read_json("https://api.coinmarketcap.com/v1/ticker/")

# Printing out the first 10 lines
crypto.head(10)
```

```
#filtering only the features we require
crypto = crypto[['rank', 'name', 'market_cap_usd', 'symbol']]

#selecting the top 10
top10 = crypto.head(10)

#renaming the columns
top10 = top10.rename(columns={'rank':'Rank', 'name': 'Name', 'market_cap_usd':'MarketCapUSD', 'symbol':'Symbol'})

#printing top10 cryptocurrencies based on their market capitalisation
top10
```

```
#importing seaborn library for plotting
import seaborn as sns

#selecting figure size for the plot
plt.figure(figsize=(10,6))

#choosing barplot and columns to plot
sns.barplot(top10.Symbol, top10.MarketCapUSD)

#x axis labelling
plt.xlabel('Name', color=black)

# y axis labelling
plt.ylabel('US Dollars in Billions', fontsize=12)

# title labelling
plt.title('TOP 10 Cryptocurrencies by Marktet Capitalisation')

#switching off ploting axis 
plt.axis('off')

#displaying the plot
plt.show()
```

```
# converting the dataframe top10 into a csv file
t = top10.to_csv('top10.csv')

#importing files module from google collaboratory
from google.colab import files

#dowloading the top10 csv file
files.download('top10.csv')
```
